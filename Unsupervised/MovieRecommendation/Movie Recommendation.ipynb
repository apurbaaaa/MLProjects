```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df = pd.read_csv("movies.csv")
```

```python
df.head()
```

```python
selected_features = ['genres','keywords','tagline','cast','director']
print(selected_features)
```

```python
for feature in selected_features:
  df[feature] = df[feature].fillna('')
```

```python
combined_features = df['genres']+' '+df['keywords']+' '+df['tagline']+' '+df['cast']+' '+df['director']
```

```python
vectorizer = TfidfVectorizer()
```

```python
features = vectorizer.fit_transform(combined_features)
```

```python
print(features)
```

```python
#finding similarity score using cosine similarity
similar = cosine_similarity(features)
```

```python
similar.shape
```

```python
#getting movie name

movie = input("Enter movie name: \n")
```

```python
#list with all the movie names in the dataset
list_of_all_movie_titles = df['title'].tolist()
```

```python
 #finding matches(recommendations)
find_match = difflib.get_close_matches(movie, list_of_all_movie_titles)
```

```python
find_match
```

```python
close_match = find_match[0]
```

```python
close_match
```

```python
index_of_recommended_movie = df[df.title == close_match]['index'].values[0]
```

```python
index_of_recommended_movie
```

```python
#getting a list of similar movies
similarity_score = list(enumerate(similar[index_of_recommended_movie]))
```

```python
similarity_score
```

```python
sorted_similar_movies = sorted(similarity_score, key = lambda x:x[1], reverse = True) 
print(sorted_similar_movies)
```

```python
print('Movies suggested for you : \n')

i = 1

for movie in sorted_similar_movies:
  index = movie[0]
  title_from_index = df[df.index==index]['title'].values[0]
  if (i<30):
    print(i, '.',title_from_index)
    i+=1
```
