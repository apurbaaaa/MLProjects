```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('Sonar.csv', header=None)
```

```python
df.head()
```

```python
df.groupby(60).mean()
```

```python
X = df.drop(columns=60, axis=1)
y = df[60]
```

```python
X.head()
```

```python
y.head()
```

```python
# Data to plot
labels = ['Mine', 'Rock']
occurrences = [111, 97]  # Occurrences of mine and rock
colors = ['gray', 'brown']  # Custom colors for each

# Create the pie chart
plt.pie(occurrences, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=90)

# Equal aspect ratio ensures that pie is drawn as a circle
plt.axis('equal')

# Display the chart
plt.title('Occurrences of Mine vs Rock')
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, stratify = y, random_state = 42)
```

```python
y_test.value_counts(0)
```

```python
# Data to plot
labels = ['Mine_for_test', 'Rock_for_test']
occurrences = [22, 20]  # Occurrences of mine and rock
colors = ['gray', 'brown']  # Custom colors for each

# Create the pie chart
plt.pie(occurrences, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=90)

# Equal aspect ratio ensures that pie is drawn as a circle
plt.axis('equal')

# Display the chart
plt.title('Occurrences of Mine vs Rock')
plt.show()
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train, y_train)
```

```python
X_train_prediction = model.predict(X_train)
```

```python
training_data_accuracy = accuracy_score(X_train_prediction, y_train)
```

```python
training_data_accuracy
```

```python
X_test_prediction = model.predict(X_test)
```

```python
test_data_accuracy = accuracy_score(X_test_prediction, y_test)
```

```python
test_data_accuracy
```
