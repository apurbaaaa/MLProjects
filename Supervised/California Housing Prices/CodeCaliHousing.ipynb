```python
dataset["income_cat"] = pd.cut(dataset["median_income"],
                                   bins=[0., 1.5, 3.0, 4.5, 6., np.inf],
                                    labels=[1, 2, 3, 4, 5])
```

```python
dataset["income_cat"].hist()
```

```python
dataset.head() #income_cat included in dataset
```

```python
from sklearn.model_selection import StratifiedShuffleSplit
split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42) 
for train_index, test_index in split.split(dataset, dataset["income_cat"]):
        strat_train_set = dataset.loc[train_index]
        strat_test_set = dataset.loc[test_index]
```

```python
strat_test_set.hist(bins=50, figsize=(20,15)) #as we can see the split in the median income is the same despite have only 20% of the dataset
plt.show()#this would not have happened if 
```

```python
for set_ in (strat_train_set, strat_test_set): #removing income_cat so that dataset is restored to original state
    set_.drop("income_cat", axis=1, inplace=True)
```

```python
train_set = strat_train_set.copy()
train_set.head()
```

```python
train_set.plot(kind="scatter", x="longitude" , y="latitude", alpha = 0.11) #alpha blurs out, shows high density of data points
#we can see this map is of california
```

```python
train_set.plot(kind="scatter", x="longitude", y="latitude", alpha=0.4,
    s=train_set["population"]/100, label="population", figsize=(10,7),
    c="median_house_value", cmap=plt.get_cmap("jet"), colorbar=True,
)
plt.legend()
```
